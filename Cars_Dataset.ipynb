Here, The data of different cars is given with their specifications. This data is available as a CSV file. Now using pandas I'm going to analyze this dataframe.

```python
import pandas as pd
```

```python
car = pd.read_csv('Cars_data.csv')
car
```

Instruction (For data cleaning)
- Find all `Null` value in the dataset.If there is any null value in any column, then fill it with the mean of that column

```python
car.isnull().sum()
```

```python
categorical_columns = car.select_dtypes(exclude=['number']).columns  # Excludes numeric columns
for col in categorical_columns:
    car[col].fillna(car[col].mode()[0], inplace=True)  # mode()[0] gets the most common value
print(car.isnull().sum())
```
